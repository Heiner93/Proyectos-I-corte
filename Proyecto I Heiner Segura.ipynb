```python
lista=[[1,2,3],[2,4,6],[2,4,5],[2,4,4]]
lista
```

**Verificar elementos en las listas**

```python
d_lista= len (lista)

for n_lista in lista:
  if len(n_lista)!=d_lista:
    print ('las filas tienen diferentes datos')
    break
else:
  print('las filas tienen mismo número de datos')
```

**Acceder a una fila**

```python
lista[2]
```

**Acceder a una columna**

```python
Columna_revisar= 2

for columna in lista:
  print(columna[Columna_revisar])
```

**Agregar una columna**

```python
nueva_columna=9

for columna in lista:
  columna.append(nueva_columna)

print(lista)
```

**Eliminar una columna**

```python
columna_eliminar= 1

for columna in lista:
  del columna[columna_eliminar]

print(lista)
```

**Agregar una fila**

```python
lista.append ([4,5,7])
```

**Eliminar una fila**

```python
fila_eliminar=1 
lista.pop(fila_eliminar)
```

**Operaciones con escalar**

```python
Operacion=2

for fila in lista:
  for i in range(len(fila)):
    fila[i] *=Operacion

print (lista)    
```

**Operaciones aritmeticas básicas entre columnas**

```python
n_columna1=1
n_columna2=2
nueva_operacion=[]

for fila in lista:
  operacion = fila[n_columna1] + fila[n_columna2]
  nueva_operacion.append(operacion)


print(nueva_operacion) 
```
