**Varianza**

```python
lista2 = []
for i in lista:
  lista2.append((float((i-media)))**2)

Varianza = sum(lista2)/len(lista2)
Varianza
```

**Desviación estandar**

```python
desviacion = Varianza**(0.5)
desviacion
```

**Coeficiente de variación**


```python
coeficiente= desviacion/(media)

coeficiente
```
